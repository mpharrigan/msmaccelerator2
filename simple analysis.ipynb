Now lets plot the equilibrium populations from each round

```python
import glob
import numpy as np
import itertools
import matplotlib.pyplot as pp
import brewer2mpl

models = []
for fn in glob.glob('models/*.npz'):
    models.append(np.load(fn))
# sort by the number of trajectories included in the model
models = sorted(models, key=lambda m: len(m['trajs']))
```

```python
pp.figure(figsize=(9, 9))

colors = itertools.cycle(brewer2mpl.get_map('Paired', 'Qualitative', 9).mpl_colors)
for i, m in enumerate(models):
    centers = m['centers']
    pops = m['populations']

    subplot = pp.subplot(3, 3, i+1)
    pp.scatter(centers[:,0], centers[:,1], color=colors.next(),
               edgecolor='k', s=10000*pops, label=str(i))
    pp.xlim(-1, 10)
    pp.ylim(-1, 10)
    pp.legend()
```

Now lets plot the first five timescales with each round

```python
import scipy.sparse
from msmbuilder.msm_analysis import get_eigenvectors

pp.figure(figsize=(6,9))

for i, m in enumerate(models):
    t_matrix = m['t_matrix'].item()
    eigenvalues = get_eigenvectors(t_matrix, 5+1)[0]
    timescales = -1/np.log(eigenvalues[1:])
    
    print 'Round %d: %s' % (i, timescales)

    pp.subplot(3,3, i+1)
    for t in timescales:
        pp.plot([0, 1], [t,t], color='b')
        pp.xticks([])
        pp.ylim(5, 20)
```

Let's try plotting the eigenprocesses themselves

```python
import matplotlib.cm

which_eigenprocess = 1

pp.figure(figsize=(10,9))
for i, m in enumerate(models):
    t_matrix = m['t_matrix'].item()
    centers = m['centers']
    eig_vectors = get_eigenvectors(t_matrix, which_eigenprocess+1)[1]
    eigenprocess = eig_vectors[:,which_eigenprocess]
    
    pp.subplot(3,3,i+1)
    pp.scatter(centers[:,0], centers[:,1], c=eigenprocess,
               s=75, cmap=matplotlib.cm.RdBu_r)
    pp.xlim(-1, 10)
    pp.ylim(-1, 10)
    pp.colorbar()
```
